1. Write a Python program that takes an integer as input and calculates the
sum of its digits


```python
number = input("Enter an integer having 2 or more digits: ")
total = 0

for n in number:
  total += int(n)

print('Sum of the digits in',number,'is',total)
```

2. Create a Python program to calculate the factorial of a given number using
a function.

```python
def fact(n):
  if n == 1 or n == 0:
    return 1
  else:
    return n * fact(n - 1)

number = int(input('Enter a number: '))
result = fact(number)
print('Factorial of',number,'is',result)
```

3. Develop a Python program that checks whether a given word or phrase is a
palindrome.

```python
word = input('Enter a word: ')

if word == word[::-1]:
  print(word,'follows a palindrom sequence')
else:
  print(word,'does not follows a palindrome sequence')
```

4. Create a Python program that takes a sentence as input and counts the
number of vowels and consonants.

```python
sentence = input('enter a sentence: ')
vowels = 'aeiou'
vowels_count = 0
consonants_count = 0

for v in range(len(sentence)):
  if sentence[v].isalpha():
    if sentence[v] in vowels:
      vowels_count += 1
    else:
      consonants_count += 1

print('Number of vowels in the sentence -',sentence,'- is',vowels_count)
print('Number of consonants in the sentence -',sentence,'- is',consonants_count)

```

5. Develop a Python program that takes two lists as input and returns a new
list containing the common elements between them.

```python
list1 = input('Enter your first list: ')
list2 = input('Enter your second list: ')
list3 = []
print('Your first list is',list(list1))
print('Your second list is',list(list2))

for item1 in list1:
  for item2 in list2:
    if item1 == item2 and item1 not in list3:
      list3.append(item1)

print('New list with common elements from first and second list is',list3)
```

6. Write a Python program that takes a string as input and counts the
frequency of each character in the string.

8. Write a Python program that checks if a given number, is a prime number.


```python
number = int(input('Enter a number: '))

if number <= 1:
  print(number,'is not a pime number')
else:
  if number > 1:
    for n in range(2,number):
      if number%n == 0:
        print(number,'is not a prime number')
        break
    else:
      print(number,'is a prime number')


```

9. Write a Python program to generate the Fibonacci sequence up to a
specified number of terms.

```python
number  = int(input('Enter a number: '))

count = 0
n1 = 0
n2 = 1

while count < number:
  print(n1,'')
  n3 = n1 + n2
  n1 = n2
  n2 = n3
  count += 1
```

10. Write a python program to find the roots of a quadratic equation

```python
import cmath

a = int(input('Enter number a: '))
b = int(input('Enter number b: '))
c = int(input('Enter number c: '))

d = cmath.sqrt(b**2 - 4*a*c)

r1 = (-b + d) / (2*a)
r2 = (-b - d) / (2*a)

print("The roots are:", r1, "and", r2)
```
