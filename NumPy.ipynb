# Explore Numpy

```python
# creating 2 lists
list1 = [1,2,3,4]
list2 = [5,6,7,8]
print(type(list1),type(list2))
```

```python
# creating arrays
array1 = np.array([2,4,6,8])
array2 = np.array([1,3,5,7])
print(type(array1),type(array2))
```

```python
# adding two lists
sum_list = list1 + list2
print(sum_list)

# adding two arrays
sum_array = array1 + array2
print(sum_array)
```

```python
# element wise addition in 2 lists
sum_list1 = []
for m,n in zip(list1, list2):
  sum_list1.append(m+n)
print(sum_list1)
```

```python
# adding arrays with different size
array1 = np.array([2,4,6,8,10])
array2 = np.array([1,3,5,7])

# adding two arrays
sum_array = array1 + array2
print(sum_array)
```

```python
# Example
item_prices = [100,200,250]
discount = 15
discounted_price = []

# Calculating 15% discount for each item
for price in item_prices:
  discount_amount = price * discount / 100
  price_after_discount = price - discount_amount
  discounted_price.append(price_after_discount)

print("Discounted prices are (Using Lists) :",discounted_price)

# If the above example is done using arrays
item_prices_array = np.array(item_prices)
type(item_prices_array)

# Calculating discount amount for each element in the array
discount_amount_array = item_prices_array * discount / 100
discounted_price_array = item_prices_array - discount_amount_array
print("Discounted prices are (Using arrays) :",discounted_price_array)
```

```python
# Comparing pythin list Vs numpy array
range1 = range(10000000)
range2 = range(10000000)

# converting the range to an array
range_array1 = np.array(range1)
range_array2 = np.array(range2)

# Using python functions
start_time = time.time()
sum_list1 = []
for m,n in zip(range1, range2):
  sum_list1.append(m+n)

end_time = time.time()

# Finding execution time
time_taken = end_time - start_time
print('Execution time for python = ',time_taken)

# using array operations
start_time = time.time()

sum_array = range_array1 + range_array2

end_time = time.time()

# finding time taken
time_taken = end_time - start_time
print('Execution time for array = ',time_taken)

# print(time_taken)
```
