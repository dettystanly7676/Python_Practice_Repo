# Q1. Write a function that return the range (maximum and minimun) of a given list of numbers

```python
def find_range(input_list):
  return max(input_list), min(input_list)

sample_list = [33,5,77,2,44,90,88,1,56,300]
print('Your list is:',sample_list)
print('maximum and minimum in the given list is',find_range(sample_list))
```

```python
def find_range(input_list):
  min = sample_list[0]
  max = sample_list[0]
  for item in input_list:
    if item < min:
      min = item
    if item > max:
      max = item
  return min, max

sample_list = [33,5,77,2,44,90,88,1,56,300]
print('Your list is:',sample_list)
print('Minimum and Maximum values in the list are:',find_range(sample_list))
# range = find_range(sample_list)
# print(range)
```

## Q2. Write function that returns mean value of a list of numbers

```python
def mean_fun(input_list):
  return sum(input_list) / len(input_list)

sample_list = [33,66,8,2,4,9,10]
print('Your list is:',sample_list)
print('Mean value of the given list is:',mean_fun(sample_list))
```

```python
# funtion to find the sum from a list of numbers
def find_mean(input_list):
  total = 0
  for number in input_list:
    total += number
  mean_value = total / len(input_list)
  return mean_value

number_list = [3,6,8,4,6,1,4]
print('Mean value of the given list is:',find_mean(number_list))
```

# Q3. Write a function that returns the median value of a list of numbers

```python
def median_fun(input_list):
  new_list = sorted(input_list)
  print('Sorted list',new_list)
  mid = int(len(new_list) + 1)
  mid_index =  int(mid /2)
  return new_list[mid_index - 1 ]


sample_list = [4,77,2,66,9,8,6]
print('Your list is:',sample_list)
print('Median value of the above list is:',median_fun(sample_list))
```

```python
# funtion to find the median value from a list of numbers
def find_median(input_list):
  sorted_list = sorted(input_list)
  print('sorted list:',sorted_list)
  mid = len(sorted_list) // 2


  if len(sorted_list)%2 != 0:
    return sorted_list[mid]
  else:
    mid_value = (sorted_list[mid] + (sorted_list[mid] - 1)) // 2
    return mid_value

number_list = [2,4,1,6,3,7]
print(find_median(number_list))
```

# Q6. Write a function that returns the lower bound of a list of numbers

```python
def lower_bound_fun(input_list):
  min_value = input_list[0]

  for number in input_list:
    if number < min_value:
      min_value = number
  return min_value

sample_list = [2,4,77,0,12,56,33]
print('Your list is:',sample_list)
print('Lower bound value in the above list is:',lower_bound_fun(sample_list))
```

##Q8. Write a function to return standard deviation of a list of numbers

```python
def stand_dev(input_list):
  mean = sum(input_list)
  new_list = []
  for item in input_list:
    new_list.append((item-mean)**2)
  return new_list



sample_list = [2,4,6,3,8,1,5,9]
print('Your list is:',sample_list)
print('Standard Deviation of the list numbers given above is:',stand_dev(sample_list))
```
